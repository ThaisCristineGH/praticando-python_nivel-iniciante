As variáveis são fundamentais em qualquer linguagem de programação, pois permitem o armazenamento e a manipulação de dados. No Python, a tipagem é dinâmica, o que significa que não precisamos declarar o tipo da variável explicitamente; ele é inferido automaticamente pelo interpretador com base no valor atribuído. Para criar uma variável, basta definir um identificador (nome), utilizar o operador de atribuição (=) e fornecer o valor desejado, conforme a sintaxe abaixo:

```python
nome = 'Thais Cristine'
```

```python
print (nome)
```

```python
nome = 'Garrafa de Água'
```

```python
print (nome)
```

```python
nome = 'Cereja'
```

```python
print (nome)
```

💡 A dinâmica é: o nome da variável funciona como um post-it. Eu posso descolar esse post-it de um número e colá-lo em um texto, ou qualquer outro objeto, a qualquer momento. No Python, os nomes referenciam valores, e essa flexibilidade é o que chamamos de reatribuição dinâmica.

```python
resultado = 10 - 5
```

```python
print (resultado)
```

```python
resultado = 10 * 2
```

```python
print (resultado)
```

```python
resultado = resultado + 10
```

```python
print (resultado)
```

```python
caixa = 20 + 30

print (caixa)

caixa = caixa + 10

print (caixa)
```
