O comando input() é a porta de entrada para a interatividade no Python, permitindo que o programa receba dados diretamente do usuário.

```python
print('Vamos começar?')
```

```python
nome =input ('Digite o seu primeiro nome')
```

```python
print(f'Olá, {nome}!')
```

```python
idade = int(input('Quantos anos você tem? '))
```

```python
print(f'Agora já sabemos o seu nome e sua idade. Nome: {nome} e idade: {idade} anos.')
```
