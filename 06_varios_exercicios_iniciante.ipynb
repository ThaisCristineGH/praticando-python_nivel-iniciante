4 - Faça um programa que receba um número inteiro e calcule sua raiz quadrada e
exiba o resultado.

```python
numero =int(input('Digite um número inteiro para calcularmos a sua raiz quadrada:'))
```

```python
raiz = numero ** (1/2) # número ** 0.5
```

```python
print (f'A raiz quadrada de: {numero} é: {raiz}')
```
