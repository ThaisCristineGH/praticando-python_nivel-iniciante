Operação: DIVISÃO  de números inteiros (números sem casas decimais) INT

```python
print ('8 / 2 =' , 8 / 2)
```

Operação: DIVISÃO  de números decimais (números com casas decimais) FLOAT

```python
print ('6 / 5 =' , 6 / 5)
```

Operação: DIVISÃO INTEIRA - Realiza a divisão e descarta a parte decimal (retorna apenas o quociente inteiro)

💡️Dica Extra: Lembre-se que o // é o "parceiro" do % (Módulo). Enquanto o // te dá o resultado inteiro, o % te dá o que sobrou da conta.

```python
print ('10 // 3 =' , 10 // 3)
```

Operação: RESTO DA DIVISÃO - Retorna apenas o que sobrou da divisão. Muito usado para verificar se um número é par ou ímpar.

💡️Dica Extra: Lembre-se que o // é o "parceiro" do % (Módulo). Enquanto o // te dá o resultado inteiro, o % te dá o que sobrou da conta.

```python
print ('10 % 4 =' , 10 % 4)
```

Operação: POTÊNCIA

```python
print ('3 ** 4 =' , 3 ** 4)
```
