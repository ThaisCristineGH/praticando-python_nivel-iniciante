1 - Comparações lógicas:

```python
print ('Comparações lógicas')
print ('1 > 2 = ', 1 > 2) # [ > ] maior
print ('2 > 2 = ', 2 > 2) # [ > ] maior
print ('3 > 1 = ', 3 > 1) # [ > ] maior
print ('1 < 2 = ', 1 < 2) # [ < ] menor
print ('5 < 8 = ', 5 < 8) # [ < ] menor
print ('3 < 6 = ', 3 < 6) # [ < ] menor
```

```python
print ('cereja' == 'cereja') # [ == ] sinal de igualdade
print ('cereja' == 'Cereja') # [ == ] sinal de igualdade
print ('cereja' != 'banana') # [ != ] sinal diferente
print ('café' != 'café') # [ != ] sinal diferente
```
